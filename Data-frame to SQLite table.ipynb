# Load data from excel file to database 

```python
df = pd.read_excel("./HR_Employee_Salary.xlsx")
```

```python
query = """SELECT * FROM EmployeeSalaryTable"""
df1 = pd.read_sql_query(query, conn)
df1.head()
```

# Create table first then load data from CSV file

```python
conn.execute('''
DROP TABLE IF EXISTS EmployeeTable
            ''')

try:
    conn.execute('''
    CREATE TABLE IF NOT EXISTS EmployeeTable(EmployeeID text, 
                                             FirstName text, 
                                             LastName text, 
                                             Gender text,
                                             Department text)
                ''')
    
    conn.commit()
    print("Table created successfully")
except:
    print("Already table existed")

conn.close()
```
